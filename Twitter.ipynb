```python
df = pd.read_csv("C:/Users/Anıl Alkan/Downloads/tweets.csv")
df=df.head(1000)
df=df['text']
df
```

```python
all_sentences = []

for word in df:
    all_sentences.append(word)
    
#all_sentences    

lines= list()
for line in all_sentences:
    words= line.split()
    for w in words:
        lines.append(w)

print(lines)        
```

```python
#Removing Punctuation
import re
lines = [re.sub(r'[^A-Za-z0-9]+', '', x) for x in lines]
#lines

lines2=[]

for word in lines:
    if word !='':
        lines2.append(word)
```

```python
from nltk.stem.snowball import SnowballStemmer

s_stemmer= SnowballStemmer(language='english')

stem=[]
for word in lines2:
    stem.append(s_stemmer.stem(word))

stem  
```

```python
stem2 = []

for word in stem:
    if word not in nlp.Defaults.stop_words:
        stem2.append(word)
stem2        
```

```python
df = pd.DataFrame(stem2)

df= df[0].value_counts()
```

```python
df
```

```python
from nltk.probability import FreqDist

freqdoctor= FreqDist()

for words in df:
    freqdoctor[words] +=1

freqdoctor    
```
